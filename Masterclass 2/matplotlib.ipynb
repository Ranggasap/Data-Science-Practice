```python
import numpy as np
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y)
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Judul Grafik')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.subplot(122)
plt.plot(x,y, color='blue')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.show()
```
