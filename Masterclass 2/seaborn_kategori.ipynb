```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('kapal_titanic.csv')
```

```python
dataku
```

```python
dataku.describe(include='O')
```

```python
# barplot (histogram)
sns.barplot(x = 'sex', y = 'fare', data=dataku);
```

```python
sns.barplot(x = 'fare', y = 'sex', data=dataku);
```

```python
sns.barplot(x = 'sex', y = 'fare', data=dataku, estimator = sum);
```

```python
sns.barplot(x = 'sex', y = 'fare', data=dataku, estimator = np.std);
```

```python
# Count plot
sns.countplot(x='sex', data=dataku);
```

```python
sns.countplot(x='deck', hue='sex', data=dataku);
```

```python
sns.countplot(x='deck', hue='sex', data=dataku, order = ['A','B','C','D','F','G']);
```

```python
# Boxplot (box & whisker plot)
sns.boxplot(x = 'sex', y= 'fare', data=dataku);
```

```python
sns.boxplot(x = 'sex', y= 'fare',hue = 'deck', data=dataku, hue_order=['A','B','C','D','E','F','G']);
```

```python
sns.boxplot(x = 'sex', y= 'fare', data=dataku);
```

```python
# Violin Plot
sns.violinplot(x='sex', y ='fare', data = dataku);
```

```python
sns.violinplot(x = 'sex', y= 'fare',hue = 'deck', data=dataku, hue_order=['A','B','C','D','E','F','G']);
```

```python
sns.boxplot(x = 'sex', y= 'fare',hue = 'deck', data=dataku, hue_order=['A','B','C','D','E','F','G']);
```

```python
sns.violinplot(x='deck', y='fare', data= dataku, hue = 'sex',order=['A','B','C','D','E','F','G'], split=True);
```

```python
# stripplot
sns.stripplot(x='deck', y='fare',hue='sex', order=['A','B','C','D','E','F','G'], data= dataku, jitter=True);
```

```python
# swarmplot
sns.swarmplot(x='deck', y='fare',hue='sex', order=['A','B','C','D','E','F','G'], data= dataku);
```

```python
sns.violinplot(x='deck', y='fare', order=['A','B','C','D','E','F','G'], data= dataku);
sns.swarmplot(x='deck', y='fare', order=['A','B','C','D','E','F','G'],color='black', data= dataku);
```
