```python
satuNew = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                    'key2':['K0','K1','K0','K1'],
                    'A': ['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})

duaNew = pd.DataFrame({'key1' : ['K0','K1','K1','K2'],
                      'key2':['K0','K0','K0','K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
satuNew
```

```python
duaNew
```

```python
baruNew = pd.merge(satuNew,duaNew, how='left', on=['key1','key2'])
```

```python
baruNew
```

```python
baruNew2 = pd.merge(satuNew,duaNew, how='outer', on=['key1','key2'])
```

```python
baruNew2
```
